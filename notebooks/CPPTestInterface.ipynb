```c++
double printCrops(std::vector<double>& cropAreas, std::vector<double>& livestockAreas, std::vector<string>& cropStrings, std::vector<string>& livestockStrings){
    int i = 0;
    double totalCropArea = 0;
    for(double d : cropAreas){
        totalCropArea += d;
        cout << cropStrings[i++] << '\t' << d << '\n';
    }
    i=0;
    for(double d : livestockAreas){
        totalCropArea += d;
        cout << livestockStrings[i++] << '\t' << d << '\n';
    }

    cout << "\nTOTAL\t" << totalCropArea;
    return totalCropArea;
}
double totalCropArea = printCrops(cropAreas, livestockAreas, cropStrings, livestockStrings);
```

```c++
// This cell re-calculates a new set of cropAreas and livestockAreas to test run() with new values
int numCrops = cropAreas.size();
int numLivestock = livestockAreas.size();

cout << "numCrops = " << numCrops << endl;
cout << "numLivestock = " << numLivestock << endl;

std::vector<double> 
    cropPropsEA = { 0.1, 0.05, 0.1, 0.075, 0.125, 0.07, 0.06, 0.05, 0.05, 0.025, 0.05 },
    livestockPropsEA = { 0.05, 0.025, 0.075, 0.025, 0.03, 0.04 },
    cropPropsSW = { 0.0002, 0.008, 0.006, 0.02, 0.03, 0.0, 0.001, 0.004, 0.008, 0.0004, 0.02 },
    livestockPropsSW = { 0.2, 0.3, 0.0002, 0.0002, 0.002, 0.4 };

std::vector<double> cropProps = cropPropsSW;
std::vector<double> livestockProps = livestockPropsSW;

double sumProps = 0;
for(double d : cropProps) sumProps += d;
for(double d : livestockProps) sumProps += d;
cout << "sumProps = " << sumProps << endl;

for(size_t i = 0; i < cropAreas.size(); i++){
    cropAreas.at(i) = totalCropArea * cropProps.at(i);
}

for(size_t i = 0; i < livestockAreas.size(); i++){
    livestockAreas.at(i) = totalCropArea * livestockProps.at(i);
}

cout << "getLowlandArea() = " << getLowlandArea(cropAreas, livestockAreas) << '\n';
cout << "getUplandArea()  = " << getUplandArea(livestockAreas) << '\n';
cout << "maxUplandArea    = " << maxUplandArea << '\n';
```
